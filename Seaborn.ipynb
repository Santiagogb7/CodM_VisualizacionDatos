```python
import matplotlib.pyplot as plt
import seaborn as sns

pinguinos = sns.load_dataset("penguins")

pinguinos.columns = ["especie", "isla", "longitud_pico", "profundidad_pico",
                     "longitud_aleta", "masa", "sexo"]
pinguinos
```

```python
sns.set_theme(style="darkgrid", context="talk")
relacional = sns.relplot(data=pinguinos,
                        height=7,
                        aspect=1.2,
                        markers=["s", ".", "P", "d"])
relacional.set(title="Analisis pinguinos")
relacional.set_ylabels("Variables")
relacional.set_xlabels("Numero de pinguinos")

plt.show()
```

```python
# Diagramas de dispersion

print(">>>>>>Simple")

sns.relplot(data=pinguinos, 
           x="longitud_pico",
           y="longitud_aleta")
plt.show()
print(">>>>>Matiz y tamanio fijo")

sns.relplot(data=pinguinos, 
           x="longitud_pico",
           y="longitud_aleta",
           hue="masa",
           s=300)

sns.relplot(data=pinguinos, 
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
           s=300)

plt.show()
```

```python
print(">>>>>>>>Rejilla")
sns.relplot(data=pinguinos, 
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
            row="sexo",
           size="masa",
           sizes=(100,500),
           style="sexo")

plt.show()
```

```python
sns.relplot(data=pinguinos, 
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
            col="isla",
            row="sexo",
           size="masa",
           sizes=(100,500),
           style="sexo")

plt.show()
```

```python
# Graficos de linea
figura, ejes = plt.subplots(1,2,figsize=(15,5))

sns.lineplot(data=pinguinos,
           x="especie",
           y="masa",
           hue="sexo",
           style="sexo",
           ax=ejes[0])

sns.scatterplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
           s=300,
            ax=ejes[1])
plt.show()
```

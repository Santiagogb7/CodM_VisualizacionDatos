```python
import numpy as np
np.random.normal(loc=100, scale=10, size=1000).std()
```

Tres distribuciones importantes
Normal, uniforme, Exponencial

```python
datos=np.random.normal(size=1000000)
plt.hist(datos, color="lightblue", ec="black")
plt.title("Distribucion Normal")
plt.show()

datos=np.random.uniform(size=1000000)
plt.hist(datos, color="darkred", ec="black")
plt.title("Distribucion Uniforme")
plt.show()

datos=np.random.exponential(size=1000000)
plt.hist(datos, color="lightgreen", ec="black")
plt.title("Distribucion Exponencial")
plt.show()
```
